```python

import yfinance as yf

data = yf.download('AAPL', start = '2012-01-01', end='2022-01-01')


print(data.describe())


plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='Closing Price', color='blue')
plt.title('Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()


data['Daily Return'] = data['Close'].pct_change() * 100


plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Daily Return'], label='Daily Return', color='green')
plt.title('Daily Returns Over Time')
plt.xlabel('Date')
plt.ylabel('Percentage Return')
plt.legend()
plt.grid(True)
plt.show()


mean_return = data['Daily Return'].mean()
std_return = data['Daily Return'].std()
print(f"Mean Daily Return: {mean_return:.2f}%")
print(f"Standard Deviation of Daily Return: {std_return:.2f}%")


trading_days_per_year = 252
annualized_return = mean_return * trading_days_per_year
annualized_volatility = std_return * (trading_days_per_year ** 0.5)
print(f"Annualized Mean Return: {annualized_return:.2f}%")
print(f"Annualized Volatility: {annualized_volatility:.2f}%")
```
